```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_conexao = ("Driver={SQL Server};"
                "Server=LAPTOP-F06PSS2R;"
                "Database=Coursera;")

conexao = pyodbc.connect(dados_conexao)
```

```python
schools_df = pd.read_sql('SELECT * FROM Coursera.dbo.ChicagoPublicSchools', conexao)
crimes_df = pd.read_sql('SELECT * FROM Coursera.dbo.ChicagoCrimeData', conexao)
census_df = pd.read_sql('SELECT * FROM Coursera.dbo.ChicagoCensusData', conexao)
display(schools_df)
display(crimes_df)
display(census_df)
```

### Problem 1: Find the total number of crimes recorded in the CRIME table.

```python
# Por pandas
qtd_crimes = len(crimes_df)
print('N√∫mero de crimes = {}'.format(qtd_crimes))
```

```python
#por SQL
qtd_crimes_sql = pd.read_sql('SELECT COUNT(*) FROM ChicagoCrimeData', conexao)
print('N√∫mero de crimes = {}'.format(qtd_crimes_sql.iloc[0][0]))
```

### Problem 2: List community areas with per capita income less than 11000.

```python
#por pandas

com_poor_areas = census_df['COMMUNITY_AREA_NAME'][census_df['PER_CAPITA_INCOME'] < 11000]
print('Community areas with per capita income less than 11000:')
for i in com_poor_areas:
    print(i)
```

```python
#por SQL
com_poor_areas_sql = pd.read_sql('SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData WHERE PER_CAPITA_INCOME < 11000', conexao )
display(com_poor_areas_sql)
```

### Problem 3: List all case numbers for crimes involving minors?

```python
# por pandas
crimes_minor=crimes_df['CASE_NUMBER'][crimes_df['DESCRIPTION'].str.contains('MINOR')]
print(crimes_minor)
```

```python
crimes_minor_sql = pd.read_sql("SELECT CASE_NUMBER FROM ChicagoCrimeData WHERE DESCRIPTION like '%minor%'", conexao)
display(crimes_minor_sql)
```

### Problem 4: List all kidnapping crimes involving a child?(children are not considered minors for the purposes of crime analysis)

```python
# por pandas
crimes_kidnapchild = crimes_df['CASE_NUMBER'][(crimes_df['PRIMARY_TYPE'].str.contains('KIDNAPPING')) & crimes_df['DESCRIPTION'].str.contains('CHILD')]
display(crimes_kidnapchild)
```

```python
#por SQL
crimes_kidnapchild_sql = pd.read_sql("SELECT CASE_NUMBER FROM ChicagoCrimeData WHERE PRIMARY_TYPE like '%kidnapping%' AND DESCRIPTION like '%child%'", conexao)
display(crimes_kidnapchild_sql)
```

### Problem 5: What kind of crimes were recorded at schools?

```python
#por pandas

crimes_school = crimes_df[['PRIMARY_TYPE', 'DESCRIPTION']][crimes_df['LOCATION_DESCRIPTION'].str.contains('SCHOOL')]
display(crimes_school)
```

```python
#por SQL

crimes_school_sql = pd.read_sql("SELECT PRIMARY_TYPE, DESCRIPTION FROM ChicagoCrimeData WHERE LOCATION_DESCRIPTION like '%school%'", conexao)
display(crimes_school_sql)
```

### Problem 6: List the average safety score for all types of schools.
